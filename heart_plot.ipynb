```python
import pandas as pd
import numpy as np
import scanpy as sc

import anndata as ad
import os
import sys
```

```python
save_path_raw = "results/human_heart/raw"
adata_st_raw = sc.read_h5ad(save_path_raw + "/model.adata_st.h5ad")
adata_st_raw.obs["slice"]
sc.pp.neighbors(adata_st_raw, use_rep='latent', n_neighbors=30)
sc.tl.louvain(adata_st_raw, resolution=0.5)
sc.tl.umap(adata_st_raw)
sc.pl.umap(adata_st_raw, color="slice", color_map="rainbow", title = "raw STitch3D")
sc.pl.umap(adata_st_raw, color="louvain", title = "raw STitch3D")
```
